## Reading and Writng data

### Drop (delete) "unnamed" column

```python
df = df.drop("Unnamed: 0", axis=1)
df
```

### Change weekly study hours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")

df.head()
```

### Gender Distribution

```python
plt.figure(figsize=(5,5))
ax= sns.countplot(data= df , x="Gender", hue="Gender")
ax.bar_label(ax.containers[0])
plt.title("Number of Males and Females", fontweight='bold', fontsize= 16 , pad= 20)
plt.show()
```

```python
# From the above chart we have analysed that: the number of females in the data is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore": "mean", "ReadingScore" : "mean" , "WritingScore": "mean"})

print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot= True ) # annot=True: Displays numerical values in cells.

plt.title("Student Score by ParentEducation", fontweight='bold', fontsize= 16 , pad= 20)

plt.show()
```

```python
# From the above chart we have concluded that the education of the parents have a good impact on the scores
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": "mean", "ReadingScore" : "mean" , "WritingScore": "mean"})

print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1, annot= True ,cmap= "Blues" )

#annot=True: Displays numerical values in cells.

#colors customized using the cmap parameter (e.g., 'viridis', 'Blues', 'coolwarm').

plt.title("Student Score by ParentMaritalStatus", fontweight='bold', fontsize= 16 , pad= 20)
plt.show()
```

```python
# From above chart we have concluded that there is no /negligible impact on the student's score due to their parent's marital status
```

### Distribution Of Ethnic Groups

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()

groupB = df.loc[(df["EthnicGroup"] == "group B")].count()

groupC = df.loc[(df["EthnicGroup"] == "group C")].count()

groupD = df.loc[(df["EthnicGroup"] == "group D")].count()

groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["groupA","groupB","groupC","groupD","groupE"] 


my_list =[groupA["EthnicGroup"], groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

plt.pie(my_list, labels=l ,autopct="%1.2f%%") 
#autopct is a parameter used specifically in Pie Charts (via plt.pie()) to display the numerical percentage of each slice inside the chart automatically

plt.title("Distrubution Of Ehtnic Groups")
plt.show()
```

```python
plt.figure(figsize=(7,4))

ax = sns.countplot(data= df, x="EthnicGroup", hue='EthnicGroup') 

sns.move_legend(ax, "upper right")

# Label all bars
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
# Conclusion : Group C has the highest number of records in the dataset, followed by Group D and Group B.Group A has the lowest representation, showing that the data is unevenly distributed across ethnic groups.
```
